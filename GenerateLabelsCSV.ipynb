```python
import os, csv
from google.colab import drive
drive.mount('/content/drive')

path_csv = '/content/drive/MyDrive/project_new/copy/DataSet/'
path = '/content/drive/MyDrive/Project/Dataset/'

with open(path_csv + "hpss_labels.csv", 'w+') as f:
    w = csv.writer(f)
    w.writerow(["filename", "label"])
    for _, dirs, _ in os.walk(path + "HPSS"):
        for dir in dirs:
            dir_int = int(dir)-1
            if(dir_int <= 10):
                for _, _, files in os.walk(os.path.join(path,"HPSS", dir)):
                    file_counter = 0
                    for file in files:
                        w.writerow([file, dir])
                        file_counter += 1
                print("saved {} files in directory {}".format(file_counter, dir))
```
